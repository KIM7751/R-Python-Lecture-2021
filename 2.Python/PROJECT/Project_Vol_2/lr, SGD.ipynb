```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, cross_validate, GridSearchCV
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import SGDClassifier
```

```python
hotel_df = pd.read_csv('C:/workspace/R/2.Python/PROJECT/Project_Vol_2/hotel_bookings_cancel.csv')
```

```python
hotel_df
```

```python
# 데이터 전처리 children, agent

hotel_df['children'] = hotel_df['children'].fillna(0)
hotel_df['agent'] = hotel_df['agent'].fillna(-1)
hotel_df
```

```python
hotel_df = hotel_df.drop(['reservation_status'], axis =1)
```

```python
hotel_df = pd.get_dummies(hotel_df)
hotel_df
```

```python
X = hotel_df.drop(['is_canceled'], axis =1)
X
```

```python
y = hotel_df['is_canceled']
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# max_iteration
model = LogisticRegression(max_iter=10000)
```

```python
model.fit(X_train, y_train)
```

```python
print("훈련 데이터 점수 : {}".format(model.score(X_train, y_train)))
print("평가 데이터 점수 : {}".format(model.score(X_test, y_test)))
```

```python
model = make_pipeline(StandardScaler(), LogisticRegression(solver='lbfgs',max_iter=10000 ))
```

```python
model.fit(X_train, y_train)
```

```python
print("학습 데이터 점수 : {}".format(model.score(X_train,y_train)))
print("평가 데이터 점수 : {}".format(model.score(X_test,y_test)))
```

```python
# 경사하강법을 이용한 분류
model = make_pipeline(StandardScaler(), SGDClassifier(loss='log'))
```

```python
model.fit(X_train,y_train)
```

```python
print("학습 데이터 점수 : {}".format(model.score(X_train,y_train)))
print("평가 데이터 점수 : {}".format(model.score(X_test,y_test)))
```
