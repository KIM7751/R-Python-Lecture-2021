```python
# 반복문 Loop

# 1) 기본 문법: while True | False : 둘 중에 하나만 올 수 있고 F = 실행X / T = 무한 반복

# 2) 반복 조건: 관습적인 변수 i , 몇 번을 반복했는지 i에 저장해서 원하는 반복 횟수만큼 / 실제로 반복됬는지 체크해서 반복 종료
#    어떻게 반복을 멈출까? --> 원하는 만큼 T 값을 이용하고 F로 전환해 멈추면 되겠다.

# 3) 활용 --> "Hello world 0,9,18,27,36...81" 10 개를 반복해보자

i=0 # 반복문 관습적 변수인 i 초기값 설정

while i<10 :    # i가 10보다 작을때 까지 반복할게(총 10번 : 0 ~ 9)
    print('"Hello world '+ str(i*9) +'"') # 문자열과 숫자열의 합은 이뤄질 수 없으므로 숫자를 str() 처리해 문자열로 바꿔 더해주는 모습
    i+=1       # i = i + 1
```

```python
# P136 연습문제
i=0                  # 변수 초기값 설정
while i<11:          # i의 범위는 1 ~ 10
    if i%3 !=0:      # 단, i는 나눠서 나머지가 0이면 안돼
        print(i)     # 뽑아줘
    i+=1             # 변수 누적값 반복식
    
```

```python
# 리스트 내포 사용하기
a = [1,2,3,4]             # for문에 사용할 리스트
result= []                # 새로운 값을 출력할 리스트 
for num in a:             # 리스트 안에 원소들의 갯수 만큼 반복
    result.append(num*3)  # result.append를 사용해 빈리스트에 원소들*3 진행
    
print(result)             # 채워진 list 결과값 출력
```

```python
# 리스트 내포의 일반 문법 
# 표현식 for 항목 in 반복 가능 객체 if 조건
# 3의 짝수에만 3을 곱하고 싶어
result = [num*3 for num in a if num % 2 == 0] 
print(result)
```

```python
# for문을 활용한 A학급의 평균점수
A = [70, 60, 55, 75, 95, 90, 80, 80, 85, 100]
total = 0                                      # 합계 구할 주머니
for score in A:                                # 점수는 A 안에 있어
    total += score                             # 합계는 점수들을 더해줄거고
average = total/len(A)                         # 평균은 리스트의 갯수로 나눠서 구할거야
print(average)                  
```

```python
# while문을 활용한 A학급의 평균점수           # 반복할 건더기가 필요함.
A = [70, 60, 55, 75, 95, 90, 80, 80, 85, 100]           
total=0                                       #주머니
i=0                                           #반복할 건더기    
while i<len(A):                               #10번 반복할게
    total+=A[i]                               #총합은 리스트 하나하나
    i+=1                                      # 건더기 플레이~
print(total/len(A))
```

```python
# for문 홀수에만 2를 곱하여 저장하기
lis = [1, 2, 3, 4, 5]                        # List
result = []                                  # 결과값 주머니 하나 하나 다 넣어야하니깐
for num in lis:                              # 숫자는 리스트 안에 있어
    if num % 2 == 1:                         # 대신 조건은 홀수야
         result.append(num*2)                # 그리고 리스트 처리 할건데 홀수*2
print(result)

# 리스트 내포식(표현식 for 항목 in 반복 가능 객체 if 조건)

result = [num*2 for num in lis if num % 2 == 1]
print(result)
```

```python
# 입력값이 몇개야? ########################################################################주의 *매개변수는 가장 맨마지막에 작성해야한다
# 매개변수 안에 많은 수를 넣으면 그걸 모두 더해주는 함수를 만들고 싶어
# (*매개변수: 하나의 튜플을 만들어 입력 매개변수 갯수를 따로 지정하지 않아도 된다는 장점이 있음.)
def kky(*ky):             # kky 라는 함수고 매개변수는 무제한이야
    result = 0            # 결과값은 초반 0
    for i in ky:          # i는 매개변수 안에 있어 튜플이니깐
        result+=i         # 결과값은 결국 튜플안에 있는 숫자들을 모두 더한 꼴이야
    return print(result)  # 최종 결과값 출력해줘

kky(1,2,4,5)
```

```python
# 매개변수 응용 (매개변수, *매개변수)
def kim(some, *ky):                   # kim이라는 함수를 만들건데 kim(일반 매개변수, 튜플링 매개변수야)
    if some == "add":                 # 일반 매개 변수가 더하기면
        result = 0                    # 더하기 결과 초기값은 0
        for i in ky:                  # i 는 튜플링 매개변수 안에있고
            result+=i                 # 튜플링에 원소들을 모두 더해줄거야
    elif some == "mul":                ## 일반 매개 변수가 곱하기면
        result = 1                    # 곱하기 결과 초기값은 1
        for i in ky:                  # i 는 튜플링 매개변수 안에 있고
            result*=i                 # 튜플링에 원소들을 모두 곱해줄거야 
    return print(result)              # 결과값 출력
```

```python
kim("add",1,2,3,4)  
```

```python
kim("mul",1,2,3,4)
```

```python
# 함수의 결괏값은 언제나 하나다? --> 아니다

def add_mul(a,b):
    return a + b, a * b

add_mul(3,4) # (a+b, a*b)의 튜플 형태로 출력됨을 알수있다
```

```python
def add_mul2(a,b):
    return a + b
    return a * b

add_mul2(3,4) # return을 두번 쓰는 방법으로는 사용할 수 없다.
              # 왜냐 return을 닿는 순간 함수를 빠져나가기 때문
```

```python
# 함수 안에 변수는 바깥 변수와 다르다
a=1

def dif(a):
    a = a + 1
 
print(dif(a)) # 함수 안에서의 변수는 바깥 함수와 연동 되어 있지 않다.
print(a)

def dif2(a):
    a = a + 1
    return a
dif2(a)       # return을 활용해 바깥함수 a를 적용 시킨 모습을 확인할 수 있다.
```
