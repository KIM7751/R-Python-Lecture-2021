```python
## 선형회귀
# 외부 데이터 모델링 전처리 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('auto-mpg.csv', header = 0)
df.shape
df
```

```python
df = df.drop(['car_name','horsepower','origin'], axis = 1, inplace = False)
```

```python
df.shape
```

```python
# 선형 회귀 분석 모델 구축하기
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# X, y 분할하고 훈련/평가용 데이터 분할하기
y = df['mpg']
X = df.drop(['mpg'], axis=1, inplace=False)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
```

```python
# 선형회귀분석 모델 생성
lr = LinearRegression()
```

```python
# 모델 훈련
lr.fit(X_train, y_train)
```

```python
# 선형 회귀 분석: 평기 데이터에 대한 예측 수행 -> 예측 결과 y_predict 구하기
y_answer = lr.predict(X_test)
y_answer
```

```python
# 평가 데이터 결과값(y_test) 예측 결과값(y_predict) 오차 계산하여 모델 평가
mse = mean_squared_error(y_test, y_answer)
print()
rmse = np.sqrt(mse) # 구한 값의 제곱근을 계산하여 평가 지표 rmse 구함
rmse
```

```python
np.round(lr.intercept_,2)
```

```python
coef = pd.Series(data = np.round(lr.coef_, 2), index = X.columns)
coef.sort_values(ascending = False)
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
```

```python
# MinMaxscale
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(
    cancer.data, cancer.target, random_state=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train)
```

```python
X_train_s = scaler.transform(X_train)
X_test_s = scaler.transform(X_test)
```

```python
print(X_train.shape, X_train_s.shape)
```

```python
print(X_train.min(axis=0), X_train_s.min(axis=0)) # 최소값이 0 , 최대값이 1로 바뀌었음을 볼 수 있음.
print(X_train.max(axis=0), X_train_s.max(axis=0))
```

```python
print(X_test.min(axis=0), X_test_s.min(axis=0)) # fitting을 train 셋을 기준으로 진행했기 때문에 test 셋의 스케일링이 0 또는 1에 가까운 숫자로 변환됨을 알 수 있음
print(X_test.max(axis=0), X_test_s.max(axis=0)) 
```

```python
# 시각화해서 스케일링을 해보자
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
X, _ = make_blobs(n_samples=50, centers=5,
                  random_state=4, cluster_std=2)        # std 표준편차
plt.scatter(X[:,0], X[:,1]);
```

```python
X_train, X_test = train_test_split(X, random_state=5, test_size=0.1)
```

```python
fig, axes = plt.subplots(1, 3, figsize = (30, 10))
axes[0].scatter(X_train[:,0], X_train[:,1], c="r", s=150)
axes[0].scatter(X_test[:,0], X_test[:,1], c="b", s=150)
axes[0].set_title("original");

scaler = MinMaxScaler().fit(X_train)  # 모델링을 진행할때 X_train을 기준으로 fit을 주고 --> transform을 진행해야함. 
                                      # 내가 실수한점 fit_transform을 이용해 test에 따로 fit을 줘서 잘못된 스케일링이 된 것 
X_train_s = scaler.transform(X_train)
X_test_s  = scaler.transform(X_test)

axes[1].scatter(X_train_s[:,0], X_train_s[:,1], c="r", s=150) # 단위만 변하고 분포 모양은 그대로임.
axes[1].scatter(X_test_s[:,0], X_test_s[:,1], c="b", s=150)
axes[1].set_title("scaled");

scaler = MinMaxScaler().fit(X_test)
X_test_s  = scaler.transform(X_test)

axes[2].scatter(X_train_s[:,0], X_train_s[:,1], c="r", s=150) # 기준은 꼭 훈련셋으로 진행되어야함.
axes[2].scatter(X_test_s[:,0], X_test_s[:,1], c="b", s=150)
axes[2].set_title("bad");
```

```python
# 스케일링 전 모델점수
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(
    cancer.data, cancer.target, random_state=1)
from sklearn.svm import SVC
svm = SVC(gamma = 'auto').fit(X_train, y_train)
svm.score(X_test, y_test)
```

```python
# 스케일링 후 모델점수
scalar = MinMaxScaler().fit(X_train)
X_train_s = scalar.transform(X_train)
X_test_s = scalar.transform(X_test)

svm = SVC(gamma = 'auto').fit(X_train_s, y_train)
svm.score(X_test_s, y_test)
```

```python
# standard스케일링 후 모델점수
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler().fit(X_train)
X_train_s = scalar.transform(X_train)
X_test_s = scalar.transform(X_test)

svm = SVC(gamma = 'auto').fit(X_train_s, y_train)
svm.score(X_test_s, y_test)
```

```python
from sklearn.datasets import fetch_lfw_people
```

```python
people = fetch_lfw_people(min_faces_per_person=20, resize=0.7)
image_shape = people.images[0].shape
```

```python
people.images.shape
```

```python
image_shape=people.images[0].shape
```

```python
fig, axes = plt.subplots(2,5, figsize=(15,8), subplot_kw={'xticks':(),'yticks':()})
for target, image, ax in zip(people.target, people.images,axes.ravel()):
    ax.imshow(image)
    ax.set_title(people.target_names[target])
```

```python
people.target_names.shape
```

```python
import numpy as np
counts = np.bincount(people.target)
for i, (count, name) in enumerate(zip(counts,people.target_names)):
    print("{0:25}{1:3}".format(name, count), end='  ')
    if (i+1)%3==0: print()
```

```python
mask = np.zeros(people.target.shape, dtype=np.bool)
mask
```

```python
for target in np.unique(people.target):
    mask[np.where(people.target == target)[0][:50]] = 1
X_people = people.data[mask]
y_people = people.target[mask]
```

```python
X_people
```

```python
y_people
```

```python
print(X_people[0].min(), X_people[0].max())
print(X_people[1].min(), X_people[1].max())
print(X_people[2].min(), X_people[2].max())
```

```python
X_people /=255
```

```python
print(X_people[0].min(), X_people[0].max())
print(X_people[1].min(), X_people[1].max())
print(X_people[2].min(), X_people[2].max())
```

```python
from sklearn.neighbors import KNeighborsClassifier
X_train, X_test, y_train, y_test = train_test_split(X_people, y_people, stratify=y_people, random_state=0)
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
knn.score(X_test, y_test)
```

```python

knn2 = KNeighborsClassifier(n_neighbors=2)
knn2.fit(X_train, y_train)
knn2.score(X_test, y_test)
```

```python
knn3 = KNeighborsClassifier(n_neighbors=3)
knn3.fit(X_train, y_train)
knn3.score(X_test, y_test)
```
