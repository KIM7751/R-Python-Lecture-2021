```python
# 범주형 변수
# 원 핫 인코딩(가변수)
import os, mglearn
import pandas as pd
```

```python
name1 = ['age', 'workclass', 'fnlwgt', 'education', 'education-num',
        'marital-status', 'occupation', 'relationship', 'race', 'gender',
        'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'income']
data = pd.read_csv(os.path.join(mglearn.datasets.DATA_PATH, "adult.data")
                  ,header=None, index_col=False, names=name1)
data
```

```python
data = data[['age', 'workclass', 'education','occupation','gender','hours-per-week','income']]
data.columns
```

```python
# get_dummies 함수는 객체 타입(문자열 같은)이나 범주형을 가진 열을 자동으로 변환 해줌
data_dummies = pd.get_dummies(data)
data_dummies.columns
```

```python
data[:10]
```

```python
data_dummies[:10] # 데이터를 숫자형으로 변환된 것 뿐임
```

```python
features = data_dummies.loc[:, 'age':'occupation_ Transport-moving']
```

```python
X = features.values
```

```python
y = data_dummies['income_ >50K'].values
```

```python
X.shape # 원래 4개의 feature 이지만 숫자형으로 바꾸는 과정에서 해당 모양으로 바뀌었다
```

```python
y.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=0)
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
lr.score(X_test,y_test)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler , OneHotEncoder
```

```python
ct = ColumnTransformer(
    [("scaling", StandardScaler(),['age', 'hours-per-week']),
     ("onehot" , OneHotEncoder(sparse=False),
     ['workclass', 'education', 'gender', 'occupation']) ])
```

```python
data_features = data.drop("income", axis=1)
data_features[:3]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
data_features, data.income, random_state=0)
```

```python
ct.fit(X_train)
X_train_trans = ct.transform(X_train)
print(X_train_trans.shape)
```

```python
logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train_trans, y_train)
```

```python
X_test_trans = ct.transform(X_test)
```

```python
logreg.score(X_test_trans, y_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
```

```python
X, y = mglearn.datasets.make_wave(n_samples=120)
```

```python
mglearn.discrete_scatter(X,y);
```

```python
import numpy as np
line = np.linspace(-3, 3, 1000, endpoint = False).reshape(-1, 1)
line
```

```python
import matplotlib.pyplot as plt
dtr = DecisionTreeRegressor(min_samples_leaf=3).fit(X,y)
plt.plot(line, dtr.predict(line), label= "decision tree")
lr = LinearRegression().fit(X,y)
plt.plot(line, lr.predict(line), label = "linear reg.")
plt.plot(X[:,0], y, 'o')
plt.legend();
```

```python
from sklearn.preprocessing import KBinsDiscretizer
```

```python
kb = KBinsDiscretizer(n_bins=10, strategy='uniform')
kb.fit(X)
kb.bin_edges_
```

```python
X_binned = kb.transform(X)
X_binned
print(X[:10])
print(X_binned.toarray()[:10])
```

```python
kb = KBinsDiscretizer(n_bins=10, strategy='uniform', encode='onehot-dense')
kb.fit(X)
X_binned = kb.transform(X)
```

```python
line_binned = kb.transform(line)
lr = LinearRegression().fit(X_binned, y)
plt.plot(line, lr.predict(line_binned), label = "linear reg.")
dtr = DecisionTreeRegressor(min_samples_leaf=3).fit(X_binned,y)
plt.plot(line, dtr.predict(line_binned), label= "decision tree")
plt.plot(X[:,0], y, 'o')
plt.vlines(kb.bin_edges_[0], -3, 3, alpha= 0.2)
plt.legend();
```
